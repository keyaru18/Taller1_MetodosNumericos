# Algoritmo 3

```python
def alg_03(n):
    if n == 0:
        return 0
    x = 0
    y = 1
    for i in range(1, n):
        z = x + y
        x = y
        y = z
    return y
```

```python
print(alg_03(11))
```

```python
print(alg_03(84))
```

```python
print(alg_03(1531))
```

## Graficar

* El valor de la serie $fib(n)$

```python
import matplotlib.pyplot as plt

n = 11
fibs = [alg_03(i) for i in range(n+1)]
```

```python
plt.plot(range(n+1), fibs, marker='o')
plt.title("Serie de Fibonacci")
plt.xlabel("n")
plt.ylabel("fib(n)")
plt.grid(True)
plt.show()
```

```python
import math
n = 1531
fibs_log = [math.log10(alg_03(i)) if i > 0 else 0 for i in range(n+1)]

plt.plot(range(n+1), fibs_log, color='blue')
plt.title("Serie de Fibonacci (log10)")
plt.xlabel("n")
plt.ylabel("log10(fib(n))")
plt.grid(True)
plt.show()
```

* El valor del cociente 

    $\phi \rightarrow \frac{fib(n)} {fib(n-1)} \approx 1.618$ nmero ureo.

```python
n = 11

phi_values = [alg_03(i)/alg_03(i-1) for i in range(2, n+1)]
```

```python
plt.plot(range(2, n+1), phi_values, marker='o', color='orange', label='fib(n)/fib(n-1)')
plt.axhline(y=1.61803398875, color='r', linestyle='--', label='Nmero ureo  1.618')
plt.title("Aproximacin al Nmero ureo con Fibonacci")
plt.xlabel("n")
plt.ylabel("fib(n)/fib(n-1)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
phi = (1 + 5**0.5)/2 

n = 1531
n_values = list(range(2, 100)) 
phi_values = [phi * (1 - (-1/phi)**i)/(1 - (-1/phi)**(i-1)) for i in n_values]
```

```python
plt.plot(n_values, phi_values, marker='o', color='orange', label='Cociente Fibonacci (aprox)')
plt.axhline(y=phi, color='r', linestyle='--', label='Nmero ureo  1.618')
plt.title("Aproximacin al Nmero ureo con Fibonacci")
plt.xlabel("n")
plt.ylabel("fib(n)/fib(n-1)")
plt.grid(True)
plt.legend()
plt.show()
```
