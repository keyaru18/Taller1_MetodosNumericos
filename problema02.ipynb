# Algoritmo 2

Implementaci√≥n de Bubble Sort.
    
- Args:
arr (list): Lista a ordenar (debe contener elementos comparables)
        
- Returns:
list: Lista ordenada (la original es modificada in-place)

* Vector 1

```python
v1 = [3, 2, 5, 8, 4, 1] 
n = len(v1)
comparisons = 0
for i in range(n):
    swapped = False
    for j in range(1, n - i):
        comparisons += 1
        if v1[j] < v1[j - 1]:
            v1[j], v1[j - 1] = v1[j - 1], v1[j]
            swapped = True
    if not swapped:
        break
print("V_sorted:", v1)
print("N√∫mero de Comparaciones:", comparisons)
```
