# Algoritmo 4

Implememte la serie geométrica 

$$\sum_{n=1}^{\infty} \frac{1}{n}$$


```python
# Serie armónica: sumatoria desde n = 1 hasta un número grande

n = int(input("¿Cuántos términos quieres sumar? "))

suma = 0.0 

for i in range(1, n + 1):
    suma += 1 / i  

print(f"La suma de los primeros {n} términos es: {suma}")
```

## ¿A que valor converge el algoritmo 4?



El **algoritmo 4** (que implementa la serie armónica)

$\sum_{n=1}^{\infty} \frac{1}{n}$

**no converge** a ningún valor finito.

- Esta serie se llama **serie armónica**.  
- Su suma **crece sin límite** (aunque lentamente).  
- Por lo tanto, el algoritmo **diverge al infinito**.

### Ejemplo de valores parciales

| n       | Suma aproximada |
|----------|-----------------|
| 10       | 2.9289          |
| 100      | 5.1873          |
| 1000     | 7.4854          |
| 10000    | 9.7876          |

A medida que \( n \) aumenta, la suma sigue creciendo indefinidamente.


```python
import matplotlib.pyplot as plt

# Cantidad de términos a sumar
n = int(input("¿Cuántos términos quieres graficar? "))

suma = 0.0
valores = []  # aquí guardaremos las sumas parciales

# Calculamos las sumas parciales de la serie armónica
for i in range(1, n + 1):
    suma += 1 / i
    valores.append(suma)

# Graficamos
plt.plot(range(1, n + 1), valores)
plt.title("Crecimiento de la Serie Armónica")
plt.xlabel("Número de términos (n)")
plt.ylabel("Suma parcial")
plt.grid(True)
plt.show()
```
